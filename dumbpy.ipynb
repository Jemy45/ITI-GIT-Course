### 6) Please write a program which asks the user for a number of days. The program then prints out the number of seconds in the number of days given.
#### How many days? 1
#### Seconds in that many days: 86400

```python
days=input("How many days? ")
seconds=int(days)*24*60*60
print(f"There are {seconds} seconds in {days} days.")
```

### 7) Please write a program which asks for the number of students on a course and the desired group size. The program will then print out the number of groups formed from the students on the course. If the division is not even, one of the groups may have fewer members than specified

```python
s=int(input("How many students? "))  
g=int(input("Desired group size? "))
num_groups=s//g
num_groups+= (s % g) > 0
print(f"Number of groups formed: {num_groups}")
```
